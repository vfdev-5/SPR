```python
df = train_month_07
```

```python
df['logrenta'] = np.log(df[df['renta'] > 0]['renta'] + 1)
df.loc[df['logrenta'].isnull(), 'logrenta'] = -99
df = df.drop(['renta', 'antiguedad'], axis=1)
```

```python
full_stats = df.describe()
full_stats
```

```python
unique_ids   = pd.Series(df["ncodpers"].unique())
limit_people = 20000
unique_id    = unique_ids.sample(n=limit_people)
reduced_df = df[df.ncodpers.isin(unique_id)]
reduced_stats = reduced_df.describe()
reduced_stats
```

```python
res = np.abs(full_stats - reduced_stats).drop(['count', '25%', '50%', '75%']).drop(['ncodpers'], axis=1)
print np.sum(res, axis=1)
print np.sum(np.sum(res, axis=1))
res
```

```python
unique_ids   = pd.Series(df["ncodpers"].unique())
limit_people = 250000

print len(unique_ids), limit_people

counter = 200
found = False
tol = 0.1
while counter > 0 and not found:
    counter -= 1
    unique_id = unique_ids.sample(n=limit_people)
    reduced_df = df[df.ncodpers.isin(unique_id)]
    reduced_stats = reduced_df.describe()
    err = np.abs(full_stats - reduced_stats).drop(['count', '25%', '50%', '75%']).drop(['ncodpers'], axis=1)
    q = np.sum(np.sum(err, axis=1))
    print "counter : ", counter, " err = ", q
    if q < tol:
        found = True
    
assert found, "Failed to reduce"

reduced_stats
```
